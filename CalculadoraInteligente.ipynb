```python
nome_usuario = input('Por favor, insira seu nome: ')
print(f'Olá, {nome_usuario}! Bem vindo a minha calculadora inteligente.')

while True:
  num1 = input('Por favor, insira o primeiro número: ')
  num2 = input('Por favor, insira o segundo número: ')
  operacao = input('Por favor, informe a operação (+, -, /, *): ')

  if operacao == '+':
    resultado = float(num1) + float(num2)
    print(resultado)
  elif operacao == '-':
    resultado = float(num1) - float(num2)
    print(resultado)
  elif operacao == '/':
   resultado = float(num1) / float(num2)
   print(resultado)
  elif operacao == '*':
    resultado = float(num1) * float(num2)
    print(resultado)
  else:
    print('Operação inválida')

  repetir = input('Deseja repetir a operação? (s/n): ')
  if repetir != 's':
    break



```
